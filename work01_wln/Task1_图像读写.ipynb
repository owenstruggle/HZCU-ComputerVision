```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import imageio
import PIL.Image as Image
```

```python
# 使用cv2库读取图像然后保存为灰度图像
img = cv2.imread('./img/SnowGirl.jpg', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('./result/gray_SnowGirl.jpg', img)

plt.imshow(img, cmap='gray')
plt.axis('off')
```

```python
img = imageio.imread('./img/SnowGirl.jpg')

# 保存为 png 和 tiff 格式
imageio.imsave('./result/SnowGirl.png', img)
imageio.imsave('./result/SnowGirl.tiff', img)
```

```python
# 使用 PIL 读入图片
img = Image.open('./img/SnowGirl.jpg')

# 将图片转化为数组
arr = np.array(img)
print(arr.shape)

# 将数组转为图片
img = Image.fromarray(arr)
img_resize = img.resize((200, 200))
img_resize.save('./result/SnowGirl.bmp')

plt.imshow(img_resize)
plt.axis('off')
```
